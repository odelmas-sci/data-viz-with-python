```python
# Here is the setup of the dataframe. Feel free to use .info() or .head() to get a better understanding of what is inside the dataframe!

import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import style

df = pd.read_csv("books.csv", error_bad_lines=False)

years = []

for y in df['publication_date']:
    date = y.split("/")
    years.append(date[2])

df["year"] = years
df.info()
```

```python
publisher = df.groupby("publisher")["title"].count().sort_values(ascending = True)
publisher
```

With your dataframe set up, time to start creating visualizations!

```python
# Use your Python skills to create a visualization here.
df_top10 = pd.DataFrame(publisher.nlargest(10))

df_top10.plot.barh(legend = False, color = "skyblue", edgecolor = "blue")
plt.title("Top 10 Publishers", color = "black")
plt.xlabel('Number of Books',color = 'black')
plt.ylabel('Publishers',color = 'black')
plt.xticks(color = 'black')
plt.yticks(color = 'black')
plt.figure(figsize = (18,18))
plt.show()
```

```python
# Use your Python skills to create a visualization here.
df_top50 = df.groupby("publisher")["title"].count().sort_values()[-10:]
df_top50
df_top50_2 = pd.DataFrame(df_top50)
df_top50_2
bar_plot = sb.barplot(data = df_top50_2, x = df_top50_2.index, y = "title")
bar_plot
bar_plot.tick_params(axis='x', rotation=45)
bar_plot.set(xlabel='Publishers', ylabel='Book Counts', title = "Top 10 Publishers")
```
